# load the dataset

```python
import pandas as pd
import numpy as np
df = pd.read_excel('/content/drive/MyDrive/cardio_train.xlsx')

df['age'] =np.round(df['age'] / 365.25).astype(int)
```

model

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


X = df[['age', 'gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active']]
y = df['cardio']
# Convert X to a NumPy array
X_array = X.values

# Select the top 5 features based on mutual information
selector = SelectKBest(score_func=mutual_info_classif, k=5)
X_selected = selector.fit_transform(X_array, y)

# Get the selected feature indices
selected_indices = selector.get_support(indices=True)
selected_features = X.columns[selected_indices].tolist()
#age, weight, ap_hi, ap_lo, cholesterol

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Initialize a Random Forest classifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=10)

# Train the Random Forest classifier
rf_clf.fit(X_train, y_train)

```

basic ui

```python
import ipywidgets as widgets
from IPython.display import display

# Define the input widgets
age_widget = widgets.IntSlider(value=30, min=0, max=100, step=1, description='Age:')
weight_widget = widgets.IntSlider(value=70, min=0, max=300, step=1, description='Weight:')
ap_hi_widget = widgets.IntSlider(value=120, min=50, max=250, step=1, description='ap_hi:')
ap_lo_widget = widgets.IntSlider(value=80, min=10, max=150, step=1, description='ap_lo:')
cholesterol_widget = widgets.Dropdown(options=[('Low', 1), ('Medium', 2), ('High', 3)], value=1, description='Cholesterol:')

# Create a button widget
button = widgets.Button(description='Predict')

# Output widget to display the prediction
output = widgets.Output()

# Define the function to make a prediction
def predict_clicked(b):
    age = age_widget.value
    weight = weight_widget.value
    ap_hi = ap_hi_widget.value
    ap_lo = ap_lo_widget.value
    cholesterol = cholesterol_widget.value
    # Perform prediction using the input values
    prediction = rf_clf.predict([[age, weight, ap_hi, ap_lo, cholesterol]])
    with output:
        output.clear_output()
        if(prediction==0):
          print('Predicted CVD Risk: No')
        else:
          print('Predicted CVD Risk: Yes')

# Link the button to the predict_clicked function
button.on_click(predict_clicked)

# Display the widgets
display(age_widget, weight_widget, ap_hi_widget, ap_lo_widget, cholesterol_widget, button, output)
```
